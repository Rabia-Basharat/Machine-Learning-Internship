```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
X, y = datasets.make_regression(n_samples=100, n_features=1, random_state=4, noise=30)
# X = 2 * np.random.randn(100, 1)
# y = 6 * X ** 2 + 3*X + np.random.randn(100, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
# print(X[0:5], X2[0:5])
X_train[:5]
```

```python
plt.scatter(X, y)
```

```python
class LinearRegression:
  def __init__(self, alpha=0.001, iterations=10000):
    self.learning_rate = alpha
    self.n_iterations = iterations
    self.weights = None
    self.bias = None
    self.errors = []
    self.degree = 2
    self.weights_list = []

  def fit(self, X, y):
    n_samples, n_features = X.shape
    self.weights = np.zeros(n_features)
    self.bias = 0
    for _ in range(self.n_iterations):
      y_predicted = np.dot(X, self.weights) + self.bias
      dw = (1/n_samples) * np.dot(X.T, (y_predicted - y))
      db = (1/n_samples) * np.sum(y_predicted - y)

      self.weights = self.weights - self.learning_rate * dw
      self.bias = self.bias - self.learning_rate * db
      self.errors.append(np.mean((y - y_predicted) ** 2))
      self.weights_list.append(np.average(self.weights))

    plt.plot(self.errors, range(0, len(self.errors)))
    # plt.xlabel('Iterations')
    # plt.ylabel('MSE')
    plt.show()

  def predict(self, X):
      return np.dot(X, self.weights) + self.bias
```

```python
model = LinearRegression(alpha=0.003)
model.fit(X_train, y_train)
prediction = model.predict(X_test)
prediction
```

```python
def error(y, y_predicted):
  return np.mean((y - y_predicted) ** 2)
```
